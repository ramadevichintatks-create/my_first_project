# loop control statements

```python
# loop control statements
'''Until now, you've learned how to repeat code using loops.

Sometimes, you want to:

1)Stop the loop immediately.
2)Skip only one iteration.
3)Leave a placeholder for future code.'''

#Python provides three keywords:

1)break
2)continue
3)pass
```

```python
#1) Break (stops everything)
''' break immediately stops the loop.'''
for i in range(1,11):
    if i == 6:
        break
    print(i)    
```

```python
#2) CONTINUE (skips)
for i in range(1,11):
    if i == 3:
        continue
    print(i)    
```

```python
# PASS 
'''Sometimes you haven't written the code yet.
Python normally gives an error if a block is empty.
do nothing for now
```

```python
#Print numbers from 1 to 10, but stop when you reach 8.
count = 1
while count <= 10:
    if count == 9:
        break
    print(count)
    count += 1
    
```

```python
def multiply(*args): # tuple (1,2) -> 1*2 ,(2,3,4) -> 2*3*4 , (4) -> 4
    product = 1
    for i in range(len(args)):
       product = product * args[i] 
    #    print(f'product value is {product} and i value is {i}')
    return product

print(multiply(4,2,10,10,12))
# (4,3,2,1) -> 4*3*2*1 multiplicative identity -> 1
```

```python
# recursion :  A function that calls itself
# it must have a base case (a condition where it stops calling itself) 
# otherwise it'll recurse forever like an infinite loop or we get a recursion error.
# factorial? 2! = 2*1 ; 3! = 3*2*1; 0! = 1
# n! = n*(n-1)*(n-2)*(n-3)*...*1

def factorial(num):
    if (num==1 or num==0):
        return 1
    else:
        return (num*factorial(num-1))


num = int(input('enter a number for which you want the factorial?'))
print(f'{num} is the number you gave')
print('calculating factorial....')
print(f'Factorial of {num} is {factorial(num)}')
```
